```python
dispo_alim = pd.read_csv("Disponibilite alimentaire.csv", sep = ";")
efficacite_loi = pd.read_csv("Efficacité loi.csv", sep = ";")
essor_sect_priv = pd.read_csv("Essor secteur privé.csv", sep = ";")
pib_hab = pd.read_csv("PIB par habitant.csv", sep = ";")
population = pd.read_csv("Population.csv", sep = ";")
stabilite_pol = pd.read_csv("Stabilité politique.csv", sep = ";")
transports = pd.read_csv("Transports.csv", sep = ";")
```

### b) Transformation et pivot des données

```python
dispo_alim["Produit & élément dispo alimentaire"] = dispo_alim["Élément"] + " : " + dispo_alim["Produit dispo alimentaire"]

dispo_alim = dispo_alim.pivot_table(index = ["Pays", "Année"], 
                                          columns = "Produit & élément dispo alimentaire", 
                                          values = "Valeur produit dispo alimentaire")

dispo_alim = dispo_alim.reset_index().rename_axis(None, axis = 1)

dispo_alim.head()
```

### c) Traitement des données NaN

```python
dispo_alim.isna().mean()
```

```python
dispo_alim = dispo_alim.loc[:, dispo_alim.count() / len(dispo_alim) >= 0.5].fillna(0)
dispo_alim.head()
```

# Fusions

## fusion_1 = efficacite_loi + essor_sect_priv

```python
fusion_1 = efficacite_loi_modif.merge(essor_sect_priv_modif, on = ["Pays", "Année"], how = "outer", indicator = True)
fusion_1.head()
```

```python
fusion_1[fusion_1["_merge"]!= "both"]
```

```python
fusion_1.drop(["_merge"], axis = 1, inplace = True)
```

## fusion_2 = fusion_1 + stabilite_pol

```python
fusion_1.info()
```

```python
fusion_2 = fusion_1.merge(stabilite_pol_modif, on = ["Pays", "Année"], how = "outer", indicator = True)
fusion_1.head()
```

```python
fusion_2[fusion_2["_merge"]!= "both"]
```

```python
fusion_2.drop(["_merge"], axis = 1, inplace = True)
```

## fusion_3 = fusion_2 + transports

```python
fusion_2.info()
```

```python
fusion_2["Année"] = fusion_2["Année"].astype("float64")
```

```python
fusion_3 = fusion_2.merge(transports, on = ["Pays", "Année"], how = "outer", indicator = True)
fusion_3.head()
```

```python
fusion_3[fusion_3["_merge"] == "right_only"]
```

```python
fusion_3.drop(["_merge"], axis = 1, inplace = True)
```

## fusion_4 = fusion_3 + population

```python
fusion_3.info()
```

```python
fusion_3["Année"] = fusion_3["Année"].astype("float64")
```

```python
fusion_4 = fusion_3.merge(population, on = ["Pays", "Année"], how = "outer", indicator = True)
fusion_4.head()
```

```python
fusion_4 = fusion_4[fusion_4["Année"] != 2001]
fusion_4[fusion_4["_merge"] == "right_only"]
```

```python
fusion_4.drop(["_merge"], axis = 1, inplace = True)
```

## fusion_5 = fusion_4 + pib_hab

```python
fusion_4.info()
```

```python
pib_hab_modif["Année"] = pib_hab_modif["Année"].astype("float64")
```

```python
fusion_5 = fusion_4.merge(pib_hab_modif, on = ["Pays", "Année"], how = "outer", indicator = True)
fusion_5.head()
```

```python
fusion_5 = fusion_5[fusion_5["Année"] != 2001]
fusion_5[fusion_5["_merge"] == "right_only"]
```

```python
fusion_5.drop(["_merge"], axis = 1, inplace = True)
```

## fusion_6 = fusion_5 + dispo_alim

```python
fusion_5.info()
```

```python
fusion_finale = fusion_5.merge(dispo_alim, on = ["Pays", "Année"], how = "outer", indicator = True)
fusion_finale.head()
```

```python
fusion_finale = fusion_finale[fusion_finale["Année"] != 2001]
fusion_finale[fusion_finale["_merge"] == "right_only"]
```

```python
fusion_finale.drop(["_merge"], axis = 1, inplace = True)
```

```python
fusion_finale = fusion_finale[fusion_finale["Année"] != 2001]
```

```python
fusion_finale["Année"] = fusion_finale["Année"].astype("int64")
```

### Exportation du dataframe fusion_finale
